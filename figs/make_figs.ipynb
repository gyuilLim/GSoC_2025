```python
import matplotlib.pyplot as plt

# 데이터
methods = ["Full FT", "Freeze", "LoRA", "DoRA"]
avg_acc = [74.92, 68.08, 77.67, 78.03]

# Full FT와 Freeze를 너무 칙칙하지 않게, 은은한 파스텔 계열로 변경
colors = ["#F6CACA", "#F9E2AE", "#A3D5D3", "#7DA9C7"]  # 핑크, 파스텔 옐로우, 민트, 파스텔 블루

plt.figure(figsize=(6, 5))
bars = plt.bar(methods, avg_acc, color=colors, edgecolor="black", linewidth=0.8)

# 값 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.3, f"{yval:.1f}",
             ha='center', va='bottom', fontsize=11, color="#333333")

# 스타일 다듬기
plt.ylim(65, 81)
plt.ylabel("Avg Top-1 Accuracy (%)", fontsize=12)
plt.title("Comparison of Fine-Tuning Methods", fontsize=14, pad=15)
plt.grid(axis="y", linestyle="--", alpha=0.4)

# 블로그용 폰트/배경
plt.gca().set_facecolor("#FBFBFB")
plt.tight_layout()
plt.savefig('./acc_compare.png')
```

```python
import matplotlib.pyplot as plt

# 데이터 (평균값 사용)
methods = ["Full FT", "Freeze", "LoRA", "DoRA"]
avg_acc = [74.92, 68.08, 77.67, 78.03]  # Accuracy (%)
avg_params = [4.032, 0.548, 2.587, 2.688]  # Params (Millions)


plt.rcParams['font.family'] = 'DejaVu Sans'   # 'Helvetica', 'DejaVu Sans', 'NanumGothic' 등으로 변경 가능
plt.rcParams['font.size'] = 12          # 기본 글씨 크기
plt.rcParams['axes.titlesize'] = 14     # 제목 크기
plt.rcParams['axes.labelsize'] = 12     # 축 라벨 크기
# 색상 팔레트 (파스텔톤)
colors = ["#F6CACA", "#F9E2AE", "#A3D5D3", "#7DA9C7"]

plt.figure(figsize=(7, 5))

# 산점도 그리기
for m, acc, param, c in zip(methods, avg_acc, avg_params, colors):
    plt.scatter(param, acc, s=180, color=c, edgecolor="black", linewidth=1.2, label=m, zorder=3)
    plt.text(param, acc+0.4, m, ha="center", fontsize=11)

# 축 라벨과 제목
plt.xlabel("Parameters (Millions)", fontsize=12)
plt.ylabel("Avg Top-1 Accuracy (%)", fontsize=12)
plt.title("Accuracy vs Parameter Efficiency", fontsize=14, pad=15)

# 그리드 & 스타일
plt.grid(linestyle="--", alpha=0.5, zorder=0)
plt.gca().set_facecolor("#FBFBFB")
plt.tight_layout()
plt.savefig('./gpu_mem_compare.png')
```
